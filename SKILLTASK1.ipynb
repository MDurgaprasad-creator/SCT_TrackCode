```python
TASK1:create a bar chart or histogram to visualize the distrubution of a categorical or continous variable such as the distrubution of ages or genders in a population 
#importing important libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("task 1.csv", skiprows=4)

# Clean dataset
df = df[["Country Name", "2023"]]  # Focus on population in 2023
df = df.dropna()                   # Drop rows with missing 2023 values
df["2023"] = df["2023"].astype(int)

# Sort countries by population in 2023 (descending)
df_sorted = df.sort_values(by="2023", ascending=False)

# --- 1. Bar Chart: Top 10 Most Populated Countries in 2023 ---
top_10 = df_sorted.head(10)

plt.figure(figsize=(12, 6))
sns.barplot(
    x="2023",
    y="Country Name",
    data=top_10,
    hue="Country Name",  # set hue to avoid warning
    palette="viridis",
    legend=False         # hide the legend since y-axis already shows country names
)
plt.title("Top 10 Most Populated Countries (2023)", fontsize=14)
plt.xlabel("Population")
plt.ylabel("Country")
plt.tight_layout()
plt.show()

# --- 2. Histogram: Distribution of Country Populations in 2023 ---
plt.figure(figsize=(10, 6))
sns.histplot(df["2023"], bins=30, kde=True, color='skyblue')
plt.title("Distribution of Country Populations (2023)", fontsize=14)
plt.xlabel("Population")
plt.ylabel("Number of Countries")
plt.tight_layout()
plt.show()
```
